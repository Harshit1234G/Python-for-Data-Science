```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
x = np.array([[185, 72], [170, 56], [168, 60], [179, 68], [182, 72], [188, 77], [180, 71], [180, 70], [183, 84], [180, 88], [180, 67], [177, 76]])
```

```python
x
```

```python
plt.scatter(x[:, 0], x[:, 1])
plt.xlabel("Height")
plt.ylabel("Weight")
```

```python
c = KMeans(2)
```

```python
c.fit(x)
```

```python
prediction = c.predict(x)
```

```python
prediction
```

```python
c.cluster_centers_
```

```python
plt.scatter(x[:, 0], x[:, 1], c= prediction)
for center in c.cluster_centers_:
    plt.scatter(center[0], center[1], marker = '^', c = 'red')
```
