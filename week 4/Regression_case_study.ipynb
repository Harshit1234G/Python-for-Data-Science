## Creating dummy data for training

```python
dummy_cars_data = pd.get_dummies(cars)
```

```python
dummy_cars_data
```

## Splitting train test

```python
from sklearn.model_selection import train_test_split
```

```python
y = dummy_cars_data.loc[:, 'price']
```

```python
features = list(dummy_cars_data.columns)
features.remove('price')
x = dummy_cars_data.loc[:, features]
```

```python
import numpy as np
y = np.log(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
x_train, x_test, y_train, y_test
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
mean_squared_error(y_test, prediction)
```

## R Squared value

```python
model.score(x_test, y_test)
```

```python
model.score(x_train, y_train)
```

## Residual plot analysis

```python
residuals = y_test - prediction
```

```python
sns.regplot(x= prediction, y= residuals, fit_reg= False)
```

```python
residuals.describe()
```

#Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train, y_train)
```

```python
prediction_rf = rf.predict(x_test)
```

```python
prediction_rf
```

```python
mean_squared_error(y_test, prediction_rf)
```

```python
rf.score(x_test, y_test)
```

```python
rf.score(x_train, y_train)
```

```python
residuals_rf = y_test - prediction_rf
```

```python
residuals_rf.describe()
```
