##Replacing Service values

```python
service_train.Service = service_train.Service.replace("Yes", 1)
service_train.Service = service_train.Service.replace("No", 0)
service_test.Service = service_test.Service.replace("Yes", 1)
service_test.Service = service_test.Service.replace("No", 0)
```

```python
service_train.Service
```

## Splitting train test

```python
x_train, y_train = service_train.iloc[:, :-1], service_train.Service
```

```python
x_train, y_train
```

```python
x_test, y_test = service_test.iloc[:, :-1], service_test.Service
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
prediction = lr.predict(x_test)
```

```python
prediction
```

```python
confusion_matrix(y_test, prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, prediction)
```

## GHI report

```python
report_data = pd.read_csv("report.csv")
```

```python
report_data
```

```python
x, y = report_data.iloc[:, :-1], report_data.H_Score
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state= 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
```
